```python
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import requests
import pandas as pd
```

```python
# Create our browser scraper
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Get/check url
url = "http://www.espn.com/nba/attendance"
requests.get(url).status_code
```

```python
browser.visit(url)
```

```python
attend_df = pd.DataFrame()

for i in range(4,(2022-2004)+4+1):
    year = []
    
    if (i < 10):
        # Select each year
        browser.find_by_tag('select').click()
        browser.find_by_text(f'200{i}').click()
        
        # Create df from current page and add year column for selected year
        new_df = pd.read_html(browser.url)[0]
        
        for j in range(0, new_df.count()[1]):
            year.append(f'200{i}')
        
        new_df["Year"] = year
        
        # Append new df to attend_df and drop title rows
        attend_df = attend_df.append(new_df)
        attend_df = attend_df.drop([0,1])
        
        
    elif (i >= 10):
        browser.find_by_tag('select').click()
        browser.find_by_text(f'20{i}').click()
        
        
        new_df = pd.read_html(browser.url)[0]
        
        for j in range(0, new_df.count()[1]):
            year.append(f'20{i}')
        
        new_df["Year"] = year
        
        
        attend_df = attend_df.append(new_df)
        attend_df = attend_df.drop([0,1])
        
```

```python
clean_attend_df = attend_df.loc[attend_df[4] != '0']
clean_attend_df
```

```python
clean_attend_df = clean_attend_df.drop(columns= 0)
clean_attend_df = clean_attend_df.reset_index(drop=True)
clean_attend_df
```

```python
clean_attend_df = clean_attend_df.loc[clean_attend_df['Year'] != '2021']
clean_attend_df
```

```python
clean_attend_df.to_csv("attendance.csv")
```
